```
#Q1
import numpy as np
import matplotlib.pyplot as plt

def bisection(a,b,t):
    while b-a > t:
        m = a + (b-a)/2
        fa = np.sin(10*a) - a
        fb = np.sin(10*b) - b
        if np.sign(fa) == np.sign(fb):
            a = m
        else:
            b = m
    return a


#z1, z2, z3, z4, z5, z6, z7 are the zeros of the function f(x) = sin(10x) -x
zeros = []
t=0.00000001

a1 = -0.9
b1 = -0.8
zeros.append(bisection(a1, b1, t))

a2 = -0.8
b2 = -0.6
zeros.append(bisection(a2, b2, t))

a3 = -0.3
b3 = -0.2
zeros.append(bisection(a3, b3, t))

a4 = -0.1
b4 = 0.1
zeros.append(bisection(a4, b4, t))

a5 = 0.2
b5 = 0.3
zeros.append(bisection(a5, b5, t))

a6 = 0.6
b6 = 0.8
zeros.append(bisection(a6, b6, t))

a7 =0.8
b7 = 0.9
zeros.append(bisection(a7, b7, t))

x = np.arange(-1, 1, 0.1)
y = np.sin(10*x) - x

x2=zeros
y2=[0,0,0,0,0,0,0]

plt.plot(x,y)
plt.plot(x2, y2, 'bo')
plt.show()
```

```
#Q2.1
def z_iteration(z, x_hat):
    end = False
    errors=[]
    x_true = np.sqrt(z)
    while not end:
        temp = (z + x_hat)/(1 + x_hat)
        #print(temp)
        if abs(temp - x_hat) <= 0.001:
            end = True
        else:
            x_hat = temp
            error = x_hat - x_true
            errors.append(error)
            
    return errors

errors = z_iteration(81, 70)
#print(errors)
plt.axis([-10,100,-30,30])
plt.plot(errors)

#At each odd iteration, (z + x_hat)/(1 + x_hat) returns a number smaller than true sqrt(z),
#(i.e the first iteration returns 2.12)
#which yeilds a negative error as the true sqrt(z) value is larger than it.

#At each even iteration, (z + x_hat)/(1 + x_hat) return a number larger than sqrt(z), 
#(i.e the second iteration returns 26.59)
#which yeilds a positive error as the true sqrt(z) value is smaller than it.

#Therefore, as the x_hat value oscilates, the error also oscillates in the same phase with the x_hat due to their
#linear relationship.
```

```
#Q2.2
def z_iteration(z, x_hat):
    end = False
    errors=[]
    x_true = np.sqrt(z)
    while not end:
        temp = 0.5*(x_hat + z)/(x_hat)
        #print(temp)
        if abs(temp - x_hat) <= 0.001:
            end = True
        else:
            x_hat = temp
            error = x_hat - x_true
            errors.append(error)
            
    return errors

errors = z_iteration(81, 70)
#print(errors)
plt.axis([-10,100,-30,30])
plt.plot(errors)

#The algorithm in (a) converges faster
#convergence rate is quadratic
```

```
#Q2.3
#This function below caclulates the n-th root of z
def z_iteration(z, x_hat, n):
    end = False
    errors=[]
    x_true = np.sqrt(z)
    while not end:
        # the updaye rule for finding n-th root of z
        temp = (1/n*(x_hat*(n-1) + z/(x_hat**(n-1))))
        #print(temp)
        if abs(temp - x_hat) <= 0.001:
            end = True
        else:
            x_hat = temp
    return x_hat


x1 = z_iteration(81, 20, 4)
print(x1)

x2 = z_iteration(1000, 1200, 3)
print(x2)

#the 4th root of 81 is indeed 3 and the 3rd root of 1000 is indeed 10.
```
